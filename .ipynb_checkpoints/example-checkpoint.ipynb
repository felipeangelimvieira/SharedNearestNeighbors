```python
from SNN import SharedNearestNeighbor as SNN
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array(list(map(lambda line: [float(line.split(" ")[0]),float(line.split(" ")[1])],open("chamaleon_dataset.txt").readlines())))
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1], s = 2)
plt.show()
```

```python
snn = SNN(n_neighbors = 60,eps = 35, min_samples = 30)
y_pred = snn.fit_predict(X)
```

```python
plt.figure()
plt.scatter(X[:,0],X[:,1], s = 2, c = y_pred)
plt.show()
```
