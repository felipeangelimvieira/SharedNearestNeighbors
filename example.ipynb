```python
from SNN import SharedNearestNeighbor as SNN
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = np.array(list(map(lambda line: [float(line.split(",")[0]),float(line.split(",")[1])],open("datasets/D31.txt").readlines())))

plt.figure()
plt.scatter(X[:,0],X[:,1], s = 2)
plt.show()

snn = SNN(n_neighbors = 40,eps = 32, min_samples = 10, weighted = False)
y_pred = snn.fit_predict(X)
plt.figure()
plt.scatter(X[:,0],X[:,1], s = 2, c = y_pred, cmap = "Spectral")
plt.savefig("D31.png")
plt.show()
```
